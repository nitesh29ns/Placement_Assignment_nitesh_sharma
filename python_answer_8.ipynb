# Get all the overall ratings for each season and using plots compare the ratings for all the seasons, like season 1 ratings, season 2, and so on

```python
seson_and_rating = data[['season','average rating']]
```

```python
seson_and_rating
```

```python
season_1_rating = sum(seson_and_rating[seson_and_rating['season']==1]['average rating'].values)/len(seson_and_rating[seson_and_rating['season']==1]['average rating'].values)
season_2_rating =sum(seson_and_rating[seson_and_rating['season']==2]['average rating'].values)/len(seson_and_rating[seson_and_rating['season']==2]['average rating'].values)
season_3_rating =sum(seson_and_rating[seson_and_rating['season']==3]['average rating'].values)/len(seson_and_rating[seson_and_rating['season']==3]['average rating'].values)
season_4_rating =sum(seson_and_rating[seson_and_rating['season']==4]['average rating'].values)/len(seson_and_rating[seson_and_rating['season']==4]['average rating'].values)
```

```python
seasons_rating=[season_1_rating,season_2_rating,season_3_rating,season_4_rating]
seasons_rating
```

```python
fig, ax= plt.subplots(figsize=(10,10))
x = 'Season_1'
y = seson_and_rating[seson_and_rating['season']==1]['average rating'].values
x1 = 'Season_2'
y1 = seson_and_rating[seson_and_rating['season']==2]['average rating'].values
x2 = 'Season_3'
y2 =seson_and_rating[seson_and_rating['season']==3]['average rating'].values
x3 = 'Season_4'
y3 = seson_and_rating[seson_and_rating['season']==4]['average rating'].values 

ax1 = plt.bar(x,y)
ax2 = plt.bar(x1,y1)
ax3 = plt.bar(x2,y2)
ax4 = plt.bar(x3,y3)
plt.title("Seasons Rating")
plt.legend(seasons_rating)
plt.xlabel("Season")
plt.ylabel("Rating")
```

# Get all the episode names, whose average rating is more than 8 for every season

```python
episodes_rating_more_than_8 = []
for i in range(len(data['name'])):
    if data['average rating'][i] > 8.0:
        episodes_rating_more_than_8.append(data.iloc[i])
```

```python
len(episodes_rating_more_than_8)
```

```python
episodes_rating_more_than_8 = pd.DataFrame(episodes_rating_more_than_8)
episodes_rating_more_than_8
```

```python
episodes_rating_more_than_8['name'].value_counts()
```

```python
sns.set(font_scale = 1.2)
plt.figure(figsize=(8,8))

plt.pie(
    x=episodes_rating_more_than_8['name'].value_counts(), 
    labels=episodes_rating_more_than_8['name'],
    colors=sns.color_palette('Set3'),
    startangle=90
)


hole = plt.Circle((0, 0), 0.65, facecolor='white')

plt.gcf().gca().add_artist(hole)
plt.title(
    label="episodes_rating_more_than_8", 
    fontdict={"fontsize":16},
    pad=20)
plt.show()
```

# Get the episode name from each season with the highest and lowest rating

```python
season_rating_name = data[['season',"average rating","name"]]
```

```python
season_rating_name
```

```python
final_data = []
for i in season_rating_name['season'].unique():
    dic = {}
    season_data = season_rating_name[season_rating_name['season'] == i]
    
    dic[f"season_{i}"] = {"highest_rating":season_data['average rating'].max(),
                          "highest_rating_ep":season_data[season_data['average rating'] == season_data['average rating'].max()]['name'].values[0],
                          "lowest_rating":season_data['average rating'].min(),
                          "lowest_rating_ep":season_data[season_data['average rating'] == season_data['average rating'].min()]['name'].values[0]
    }
    final_data.append(dic)
```

```python
lowest_ep = []
lowest_rate = []
heighest_ep =[]
heighest_rate=[]
for i in range(len(final_data)):
    for j in final_data[i].items():
        #print(j[1]['lowest_rating_ep'])
        heighest_ep.append(j[1]['highest_rating_ep'])
        heighest_rate.append(j[1]['highest_rating'])
        lowest_ep.append(j[1]['lowest_rating_ep'])
        lowest_rate.append(j[1]['lowest_rating'])
```

```python
heighest_ep,heighest_rate,lowest_ep,lowest_rate
```

```python
label = ['season_1','season_2','season_3','season_4']
plt.figure(figsize=(8,8))
plt.bar([heighest_ep[0],lowest_ep[0]], [heighest_rate[0],lowest_rate[0]])
plt.bar([heighest_ep[1],lowest_ep[1]], [heighest_rate[1],lowest_rate[1]])
plt.bar([heighest_ep[2],lowest_ep[2]], [heighest_rate[2],lowest_rate[2]])
plt.bar([heighest_ep[3],lowest_ep[3]], [heighest_rate[2],lowest_rate[3]])
plt.xticks(rotation=90)
plt.legend(label, title='season');
```
